## Data cleaning


```python
# renaming columns
df=df.rename(columns={'Mat':'Matches_played','Inns':'Innings_batted',
                      'NO':'No_outs','Runs':'Runs_scored','HS':'Highest_inns_scored',
                      'BF':'Balls_faced','SR':'strike_rate'})
df
```

```python
#checking and removing duplicates
```

```python
df[df['Player'].duplicated()]
```

```python
df[df['Player'].isin(['RG Pollock (SA)','R Dravid (ICC/IND)','Mohammad Yousuf (PAK)','DPMD Jayawardene (SL)'])]
```

```python
df=df.drop_duplicates()
df
```

```python
#checking and replacing nulls
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

```python
df.isna().sum().sum()
```

```python
#splitting columns
```

```python
df['Player'].str.split(pat='(')
```

```python
df.loc[:,'Country']= df['Player'].str.split(pat='(').str[1]
```

```python
df.loc[:,'Country']=df['Country'].str.strip(')')
```

```python
df.loc[:,'Player']=df['Player'].str.split(pat='(').str[0]
```

```python
df['Span'].str.split(pat='-')
```

```python
df.loc[:,'Start_year']=df['Span'].str.split(pat='-').str[0]
```

```python
df.loc[:,'Finish_Year']=df['Span'].str.split(pat='-').str[1]
```

```python
df=df.drop(['Span'],axis = 1)
```

```python
#changing datatypes
```

```python
df.dtypes
```

```python
df['Matches_played'] = df['Matches_played'].str.strip('*')
```

```python
df['Matches_played']=df['Matches_played'].astype('int')
```

```python
df['Highest_inns_scored']=df['Highest_inns_scored'].str.strip('*')
```

```python
df['Highest_inns_scored']=df['Highest_inns_scored'].astype('int')
```

```python
df['Balls_faced']=df['Balls_faced'].str.strip('+')
```

```python
df['Balls_faced']=df['Balls_faced'].astype('int')
```

```python
df['4s']=df['4s'].str.strip('+').astype('int')
```

```python
df['6s']=df['6s'].str.strip('+').astype('int')
```

```python
df['Start_year']=df['Start_year'].astype('int')
```

```python
df['Finish_Year']=df['Finish_Year'].astype('int')
```

## PERFORMING EXPLORATORY DATA ANALYSIS

```python
df.describe()
```

```python
df['Career_length']= df['Finish_Year']-df['Start_year'] #calculating career length
```

```python
df['Career_length'].mean() #average career length of cricketers
```

```python
df=df.reset_index()
```

```python
df[df['Career_length']>10]['strike_rate'].mean()# average career length of cricketers who played over 10 years
```

```python
#players who played before 1960
df[df['Start_year']<1960]['Player'].count()
```

```python
max_country_highest_score=df.groupby(['Country'])['Highest_inns_scored'].max()#countries maximum highest score
max_country_highest_score.sort_values(ascending=False)
```

```python
max_country_highest_score.plot(kind='bar',xlabel='Country',ylabel='Highest_inns_scored')
```
